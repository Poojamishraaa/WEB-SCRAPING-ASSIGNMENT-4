```python
from bs4 import BeautifulSoup
from selenium import webdriver
driver = webdriver.Chrome()
url = "https://www.imdb.com/list/ls095964455/"
driver.get(url)
soup = BeautifulSoup(driver.page_source, 'html.parser')
driver.quit()

MoNameArr=[]
for i in soup.select('.lister-list .lister-item-content h3.lister-item-header a'):
    MoName = i.text.strip()
    MoNameArr.append(MoName)

MoNameArr[0:]
```

```python
MoYearArr=[]
for i in soup.select('.lister-list .lister-item-content h3.lister-item-header span.lister-item-year'):
    MoYear = i.text.strip().replace('(', '').replace(')', '')
    MoYearArr.append(MoYear)

MoYearArr[0:]
```

```python
MoGenreArr=[]
for i in soup.select('.lister-list .lister-item-content span.genre'):
    MoGenre = i.text.strip()
    MoGenreArr.append(MoGenre)

MoGenreArr[0:]
```
